```python
# Say anyone under 16 as child
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child, axis = 1)
```

```python
#let's use classes as categories again, but hue by person
sns.countplot('Pclass',data = titanic_df, hue = 'person')
```

```python
#visualize the data is to use FacetGrid to plot multiple Kernel Density Estimation plots on one plot

fig = sns.FacetGrid(titanic_df, hue = 'Sex', aspect = 4)
fig.map(sns.kdeplot,'Age', shade = True)

oldest = titanic_df['Age'].max()

fig.set(xlim = (0, oldest))

fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
levels = []

for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df.sort_values(by = ['Cabin'], inplace = True)
sns.countplot('Cabin', data = cabin_df, palette = 'winter_d')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.countplot('Cabin', data = cabin_df, palette = 'spring')
```

```python
titanic_df['Family'] = titanic_df.SibSp + titanic_df.Parch
```
